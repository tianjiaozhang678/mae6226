```python
import numpy
import math 
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 50
x_start, x_end = -2.0, 2.0
y_start, y_end = -1.0, 1.0
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X, Y = numpy.meshgrid(x, y)
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
    u = +strength / (2 * math.pi) * (Y - yv) / ((X - xv)**2 + (Y - yv)**2)
    v = -strength / (2 * math.pi) * (X - xv) / ((X - xv)**2 + (Y - yv)**2)
    return u, v
```

```python
def get_stream_function_vortex(strength, xv, yv, X, Y):
    psi = strength / (4 * math.pi) * numpy.log((X - xv)**2 + (Y - yv)**2)
    return psi
```

```python
u=numpy.zeros_like(X)
v=numpy.zeros_like(X)
psi=numpy.zeros_like(X)
```

```python
gamma = 5.0
a = 0.8
n = int(4.0/a +1)
x_vortex = numpy.linspace(-2.0, 2.0, n)
y_vortex = 0.0


for i in range(n):
    u_vortex, v_vortex = get_velocity_vortex(gamma, x_vortex[i], y_vortex, X, Y)
    psi_vortex = get_stream_function_vortex(gamma, x_vortex[i], y_vortex, X, Y)
    u += u_vortex
    v += v_vortex
    psi += psi_vortex


width = 10
height = (y_end - y_start) / (x_end - x_start) * width
pyplot.figure(figsize=(width, height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v,
                  density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_vortex, numpy.zeros(n), color='#CD2305', s=80, marker='o')
```
